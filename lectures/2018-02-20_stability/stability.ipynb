```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 3.1
```

```python
x = np.array([0.25])
for i in range(99):
    x = np.append(x, (a * x[i]) * (1 - x[i]))
plt.plot(x)
plt.show()
```

```python
a = .9
xx = np.linspace(start=0, stop=1, num=50)
yy = (a * xx) * (1 - xx)

start = 0.5
x = np.array([start])
y = np.array([start])
for i in range(999):
    if i % 2 == 0:
        x = np.append(x, x[-1])
        y = np.append(y, (a * x[-1]) * (1 - x[-1]))
    else:
        x = np.append(x, y[-1])
        y = np.append(y, y[-1])

plt.plot(xx, xx)
plt.plot(xx, yy)
plt.plot(x, y)
plt.show()
```
