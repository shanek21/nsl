```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.matrix([[0, 1], [-1, 0]]) # mass spring equations
I = np.matrix([[1, 0], [0, 1]]) # identity matrix
dt = 0.02 # timestep
time = np.arange(0, 10, dt)
# M = I + dt * A # forward Euler
M = np.linalg.inv(I - dt * A) # backward Euler
```

```python
x = np.mat([[1], [0]]) # initial conditions
for t in time[1:]:
    lastCol = x[:, -1]
    x = np.concatenate((x, M * lastCol), axis=1)
plt.plot(time, x.T)
plt.title('Mass and Spring')
plt.xlabel('Time')
plt.legend(['Position', 'Velocity'])
plt.show()
```
